```python
import numpy as np
import matplotlib.pyplot as plt
#Simulating Xs
x = np.random.uniform(0,100,size=50)
#Computing expected Ys
y_expected = x*3 + 100
plt.scatter(x, y_expected)
```

Simulating "measured" Ys (adding noise)

```python
y_measured =  y_expected + np.random.normal(0,50,size=50)
plt.scatter(x, y_measured)
```

an easily find regression line by library..

```python
from sklearn.linear_model import LinearRegression
estimated_relation = LinearRegression()
estimated_relation.fit(x.reshape((-1, 1)), y_measured)
plt.scatter(x, y_measured)
plt.plot(x, estimated_relation.predict(x.reshape((-1, 1))))
print(estimated_relation.coef_, estimated_relation.intercept_)
```

imulating "measured" Xs (adding noise)

```python
x_measured = x + np.random.normal(0,50,size=50)
#Plot
plt.scatter(x_measured, y_expected)
#Estimate relation
estimated_relation = LinearRegression()
estimated_relation.fit(x_measured.reshape((-1, 1)), y_expected)
plt.plot(x_measured, estimated_relation.predict(x_measured.reshape((-1, 1))))
print(estimated_relation.coef_, estimated_relation.intercept_)
```
